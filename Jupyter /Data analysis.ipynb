```python
listingcount = listingcomplete.groupby(['zipcode']).size().reset_index(name='NumOfAirbnb')
```

```python
print(listingcomplete.columns)
```

```python
listingprice = listingcomplete[['zipcode', 'price']]
def price_to_float (price):
    price = str(price)
    return float(price.strip("$").replace(",", ""))
listingprice['price'] = listingprice['price'].apply(price_to_float)
```

```python
listingmedianprice = listingprice.groupby(['zipcode'])['price'].median().reset_index(name='medianPrice')
```

```python
zipcodePriceCount = pd.merge(listingmedianprice, listingcount,  how='inner', left_on='zipcode', right_on = 'zipcode')
```

```python
print(venuescount['Zipcode'].dtype)
zipcodePriceCount['zipcode'] = zipcodePriceCount['zipcode'].astype(float)
```

```python
print(zipcodePriceCount['zipcode'].dtype)
```

```python
zipcodePriceCountVenues = pd.merge(zipcodePriceCount, venuescount,  how='inner', left_on='zipcode', right_on = 'Zipcode')
```

```python
print(zipcodePriceCountVenues)
```

```python
zipcodePriceCountVenues.rename(columns={'counts': 'VenuesCount'})
print(zipcodePriceCountVenues)
```

```python
zipcodePriceCountVenues.to_csv('ZipcodeVenuesPrice.csv')
```
